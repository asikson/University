```python
darknet_all_df = pd.read_csv("Darknet.csv", header=0)
darknet_all_df.head()
```

```python
chosen_columns = [
    "Idle Max",
    "FWD Init Win Bytes",
    "Idle Mean",
    "Idle Min",
    "Fwd Seg Size Min",
    "Subflow Fwd Packets",
    "Flow Duration",
    "Flow IAT Max",
    "Flow IAT Min",
    "Flow IAT Mean",
    "Fwd Packets/s",
    "Flow Packets/s",
    "Bwd Init Win Bytes",
    "Protocol",
    "FIN Flag Count",
    "Bwd Packets/s",
    "Fwd IAT Max",
    "Bwd Packet Length Mean",
    "Bwd Packet Length Min",
    "Fwd IAT Total",
    "Label",
    "Label.1"
]

darknet_df = darknet_all_df[chosen_columns]
darknet_df
```

```python
# Darknet = Tor + VPN

darknet_df['Label'].replace(to_replace=['Tor', 'VPN'], value='Darknet', inplace=True)
darknet_df['Label'].replace(to_replace=['Non-Tor', 'Non-VPN'], value='Non-Darknet', inplace=True)
darknet_df
```

```python
def hasMissingValues(row):
    for column, val in row.iteritems():
        if pd.isnull(val) or val == '' or (not isinstance(val, str) and m.isinf(val)):
            return True
    return False
```

```python
darknet_df = darknet_df[darknet_df.apply(lambda row: not hasMissingValues(row), axis=1)]
darknet_df
```

```python
X = darknet_df[[
    "Idle Max",
    "FWD Init Win Bytes",
    "Idle Mean",
    "Idle Min",
    "Fwd Seg Size Min",
    "Subflow Fwd Packets",
    "Flow Duration",
    "Flow IAT Max",
    "Flow IAT Min",
    "Flow IAT Mean",
    "Fwd Packets/s",
    "Flow Packets/s",
    "Bwd Init Win Bytes",
    "Protocol",
    "FIN Flag Count",
    "Bwd Packets/s",
    "Fwd IAT Max",
    "Bwd Packet Length Mean",
    "Bwd Packet Length Min",
    "Fwd IAT Total"
]]

Y = darknet_df["Label"]
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.3)
clf = DecisionTreeClassifier(splitter='best', max_depth = 7, random_state=0)
clf.fit(train_X, train_Y)

plt.figure(figsize=(250, 100))
tree.plot_tree(clf, fontsize=50,label="none")
plt.show()
```

```python
print(clf.score(train_X, train_Y))
print(clf.score(test_X, test_Y))
predictedY = clf.predict(test_X)
```

```python
forestclf = RandomForestClassifier(n_estimators=30, max_samples = 1000)
forestclf.fit(train_X, train_Y)
print(forestclf.score(test_X, test_Y))
```
